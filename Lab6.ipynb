```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('company_sales_data.csv')

# Extracting month numbers and total profit from the DataFrame
months = df['month_number'].tolist()
total_profit = df['total_profit'].tolist()

# Creating the line plot
plt.plot(months, total_profit, marker='o', color='b', linestyle='-')

# Adding labels and title
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')

# Displaying the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('company_sales_data.csv')


months = df['month_number'].tolist()


products = df.columns[1:-2]  

for product in products:
    plt.plot(months, df[product], marker='o', label=product)


plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Product Sales Data')


plt.legend()


plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('company_sales_data.csv')

# Extracting month numbers
months = df['month_number'].tolist()

# Extracting sales data for bathing soap and facewash
bathing_soap_sales = df['bathingsoap'].tolist()
facewash_sales = df['facewash'].tolist()

# Creating subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plotting sales data for bathing soap
axs[0].plot(months, bathing_soap_sales, marker='o', color='blue')
axs[0].set_title('Bathing Soap Sales Data')
axs[0].set_xlabel('Month Number')
axs[0].set_ylabel('Units Sold')

# Plotting sales data for facewash
axs[1].plot(months, facewash_sales, marker='o', color='green')
axs[1].set_title('Facewash Sales Data')
axs[1].set_xlabel('Month Number')
axs[1].set_ylabel('Units Sold')

# Adjust layout
plt.tight_layout()

# Displaying the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('company_sales_data.csv')


months = df['month_number'].tolist()

products = df.columns[1:-2]  


sales_data = []


for product in products:
    sales_data.append(df[product].tolist())


plt.figure(figsize=(10, 6))
plt.stackplot(months, sales_data, labels=products)


plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.title('Product Sales Data')


plt.legend(loc='upper left')


plt.show()
```
